```python
import torch
import clip
from PIL import Image
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
import requests

image = Image.open(requests.get("https://www.adobe.com/acrobat/hub/media_173d13651460eb7e12c0ef4cf8410e0960a20f0ee.jpg", stream=True).raw)
image_inputs = torch.stack([preprocess(image).to(device) for _ in range(2)])
image_inputs
```

```python
with torch.no_grad():
    image_feature = model.encode_image(image_inputs)
image_feature
```

```python
with torch.no_grad():
    features = model.encode_image(image_inputs)
    features /= features.norm(dim=-1, keepdim=True)
features
```
