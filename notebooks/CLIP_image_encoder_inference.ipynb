```python
import torch
import clip
from PIL import Image
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
image = preprocess(Image.open("D:/6. Datasets/SPHAR-Dataset/images/falling/casia_topdownview_p09_faint_a1_00071.jpg")).unsqueeze(0).to(device)
image, image.shape
```

```python
with torch.no_grad():
    image_feature = model.encode_image(image)
image_feature
```
